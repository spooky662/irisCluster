```python
#carregar o modelo normalizador
import numpy as np
from pickle import load
normalizador_iris = load(open('/content/drive/MyDrive/SistemasInteligentes/models/normalizador_iris.model', 'rb'))

#carregar o modelo de clusters
iris_clusters_model = load(open('/content/drive/MyDrive/SistemasInteligentes/models/iris_clusters.model', 'rb'))
```

```python
#capturar os dados de uma nova instancia
flor_num = [6.4, 2.8, 5.6, 2.1]
flor_cat = ['iris-setosa'] #vai causar stress

#normalizar os dados numericos de entrada
flor_num = normalizador_iris.transform([flor_num])
flor_num
```

```python
#estruturar os dados da nova flor
import pandas as pd
Colunas = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'Iris-setosa', 'Iris-versicolor', 'Iris-virginica']

nova_flor_data_frame = pd.DataFrame(columns = Colunas)
nova_flor_data_frame['sepal_length']=[flor_num[0][0]]
nova_flor_data_frame['sepal_width']=[flor_num[0][1]]
nova_flor_data_frame['petal_length']=[flor_num[0][2]]
nova_flor_data_frame['petal_width']=[flor_num[0][3]]
nova_flor_data_frame['Iris-setosa']=[0]
nova_flor_data_frame['Iris-versicolor']=[0]
nova_flor_data_frame['Iris-virginica']=[0]
nova_flor_data_frame
```

```python
#inferir o grupo ao qual essa flor pertence
grupo = iris_clusters_model.predict(nova_flor_data_frame)

cluster_nova_flor = iris_clusters_model.cluster_centers_[grupo[0]]

#converter o centroid em um df com a estrutura dos dados de treinamento
cluster_nova_flor = pd.DataFrame([cluster_nova_flor], columns= Colunas)
cluster_nova_flor

#desnormalizar os dados numericos
cluster_nova_flor_num = pd.DataFrame(normalizador_iris.inverse_transform(cluster_nova_flor.drop(columns=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'], axis=1)))

#converter os dados numericos em data frame
cluster_nova_flor_num = pd.DataFrame(cluster_nova_flor_num.values, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
cluster_nova_flor_num
```

```python
#desnormalizar os dados categoricos
cluster_nova_flor_cat = round(cluster_nova_flor[['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']], 0)
cluster_nova_flor_cat_desnormalizado = pd.from_dummies(cluster_nova_flor_cat, sep ='-')
cluster_nova_flor_cat_desnormalizado
```

```python
#compor o dataframe completo com os dados do centroid
centroide = cluster_nova_flor_num.join(cluster_nova_flor_cat_desnormalizado)
centroide
```
